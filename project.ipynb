- Few wines exhibit volatile acidity above 1
  
  ### Citric acid
- A significant number of wine samples contain low levels of citric acid, particularly below 0.4.
  
  ### Residual Sugar
- Most wines have residual sugar close to 2.
- The frequency decreases sharply as residual sugar increases
  
  ### Chlorides
- Most wines have chlorides close to 0.1.
- The frequency decreases sharply as chlorides increase
  
  ### Free sulfur dioxide
- Most wines have free sulfur dioxide around 20-30.
- The frequency decreases as free sulfur dioxide increases
  
  ### Total sulfur dioxide
- Most wines have total sulfur dioxide between 100-150.
- The frequency decreases as total sulfur dioxide increases
  
  ### Density
- Most wines have a density just below 1.0000 with very little variation.
  
  ### pH
- The pH level shows a normal distribution peaking at around 3.2.
  
  ### Sulphates
- Most wines have sulphates slightly before reaching a value of 1.
- The frequency decreases as sulphates increase
  
  ### Alcohol
- The alcohol content displays multiple peaks indicating varied alcohol content levels with significant peaks at approximately both ends (9% &14%).
  
  ### Quality
- The quality rating indicates two prominent peaks suggesting two common quality ratings within the dataset

 ### Scatter Diagram Analysis

- **First Scatter Plot**: Fixed acidity vs. Volatile acidity
  - Analysis:
    - The x-axis represents fixed acidity (6 to 16) and the y-axis represents volatile acidity (0.2 to 1.6).
    - Most data points are concentrated towards the lower end of both axes, indicating lower levels of both acidities.
    - A dense cluster exists around 8 to 10 fixed acidity and 0.4 to 0.8 volatile acidity, suggesting common levels for many wine samples.
    - Fewer data points are scattered at higher levels of both acidities, implying fewer samples with high acid levels.
    - Conclusion: Potential correlation observed.

- **Second Scatter Plot**: Residual sugar vs. Citric acid
  - Analysis:
    - The x-axis represents citric acid content (0 to 1) and the y-axis represents residual sugar (0 to 16).
    - Most data points are clustered between 0 to 0.4 citric acid and 2 to 6 residual sugar, indicating common levels in samples.
    - Some scattered points show higher residual sugar but lower citric acid content.
    - Conclusion: Potential correlation observed but weak due to dispersed data.

- **Third Scatter Plot**: Chlorides vs. Free sulfur dioxide
  - Analysis:
    - The x-axis represents chlorides (0.0 to 0.6) and the y-axis represents free sulfur dioxide (0 to 70).
    - Most data points cluster around 0.0 to 0.1 chlorides, spanning across the entire range of free sulfur dioxide.
    - Scattered data points are visible as chlorides increase, suggesting a potential inverse correlation with free sulfur dioxide levels.
    - Conclusion: Potential inverse correlation observed but weak due to dispersed data.

- **Fourth Scatter Plot**: Total sulfur dioxide vs. Density
  - Analysis:
    - The x-axis represents total sulfur dioxide (0 to 300) and the y-axis represents density (0.990 to 1.004).
    - Most data points cluster between 0 and 150 on the total sulfur dioxide scale, and between 0.996 and 1.002 on the density scale.
    - Fewer data points are observed as total sulfur dioxide readings increase, indicating less frequency of higher readings in the dataset.
    - Conclusion: Potential correlation observed but weak due to dispersed data.

- **Fifth Scatter Plot**: pH vs. Sulphates
  - Analysis:
    - The x-axis represents pH (2.8 to 4.0) and the y-axis represents sulphates (0.25 to 2.00).
    - Most data points cluster around pH value of 3.2 and sulphate value of approximately 0.5 to 1.
    - Fewer data points occur away from this central concentration, indicating less frequent occurrences of those combinations.
    - Conclusion: Potential correlation observed but weak due to dispersed data.

- **Sixth Scatter Plot**: Alcohol content vs. Quality of wine
  - Analysis:
    - The x-axis represents alcohol content (approximately 9 to 15) and the y-axis represents quality (3 to 8).
    - Most data points cluster between approximately 9 to 12 alcohol content and 5 to 7 quality.
    - A trend suggests higher alcohol content correlates with higher quality ratings, although exceptions exist.
    - Conclusion: Potential positive correlation observed but weak due to dispersed data.
